```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from scipy import optimize
from mpl_toolkits.mplot3d import Axes3D
plt.style.use("bmh")
matplotlib.rc('font', size=10)
```

```python
K = 0.1
M = 1.0
```

```python
# forward function
def f(p): # f(p) = m
    return M * (K * p)/(1 + K * p)

# inverse function
def g(m): # g(m) = p
    return m / (K * (M - m))
```

# Forward 1D $\kappa(p)$  

```python
p = np.linspace(0, 100)
κ = 1 / (1 + K * p)
fig, ax = plt.subplots(2, 1)
ax[0].plot(p, f(p))
ax[1].plot(p, κ)

ax[0].set_title("Langmuir Isotherm")
ax[0].set_ylabel("$f(p)$")
ax[1].set_ylabel("$κ(p)$")
ax[1].set_xlabel("$p$")
```

# 1D Inverse $\kappa(m)$ 

```python
def κ_of_m(m): # inverse kappa with respect to m
    return M / (M - m) 

def κ_of_p(p): # inverse kappa with respect to p
    return 1 + K * p
```

```python
p = np.linspace(0, 100)
m = np.linspace(0, .75)
#m = np.linspace(0, 1.67)
fig = plt.figure()
ax = plt.axes()
plt.plot(m, κ_of_m(m))
ax.set_ylabel("κ(m)")
ax.set_xlabel("m")
```
